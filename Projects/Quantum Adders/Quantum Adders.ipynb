```python
from qiskit import QuantumCircuit, assemble, Aer, execute
from qiskit.visualization import plot_histogram
```

```python
### =-=-=-=-=-= 2-qubit Half Adder =-=-=-=-=-= ###
```

```python
circuit = QuantumCircuit(4, 2)

circuit.h([0, 1])
# circuit.x(0)
# circuit.x(1)
circuit.barrier()

circuit.cx(0, 2)
circuit.cx(1, 2)
circuit.ccx(0, 1, 3)
circuit.barrier()

circuit.measure(2, 0)
circuit.measure(3, 1)

circuit.draw('mpl')
```

```python
sim = Aer.get_backend('qasm_simulator')
qobj = assemble(circuit, shots=5000)
counts = sim.run(qobj).result().get_counts()
plot_histogram(counts)
```

```python
### =-=-=-=-=-= 3-qubit Half Adder =-=-=-=-=-= ###
```

```python
circuit = QuantumCircuit(5, 2)

circuit.h([0, 1, 2])
# circuit.x(0)
# circuit.x(1)
# circuit.x(2)
circuit.barrier()

circuit.cx(0,3)
circuit.cx(1,3)
circuit.cx(2,3)
circuit.ccx(0,1,4)
circuit.ccx(0,2,4)
circuit.ccx(1,2,4)
circuit.barrier()

circuit.measure(3,0)
circuit.measure(4,1)

circuit.draw('mpl')
```

```python
sim = Aer.get_backend('qasm_simulator')
qobj = assemble(circuit, shots=5000)
counts = sim.run(qobj).result().get_counts()
plot_histogram(counts)
```
