```python
from qiskit import *
from qiskit.visualization import plot_histogram
from qiskit.tools.monitor import job_monitor
```

```python
### -=-=-=-=- Plot Histogram -=-=-=-=- ###
```

```python
# 2-qubit Bell state
circuit = QuantumCircuit(2, 2)
circuit.h(0)
circuit.cx(0, 1)

meas = QuantumCircuit(2, 2)
meas.measure(range(2), range(2))

total = circuit + meas
total.draw('mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
first_result = execute(total, simulator, shots = 1000).result()
second_result = execute(total, simulator, shots = 1000).result()
first_counts = first_result.get_counts(total)
second_counts = second_result.get_counts(total)

legend = ['First Execution', 'Second Execution']
plot_histogram([first_counts, second_counts], legend=legend)
```

```python
plot_histogram([first_counts, second_counts], legend=legend, sort='desc', figsize=(15,12),
               color=['red', 'yellow'], bar_labels=False)
```

```python
### -=-=-=-=- Plot State -=-=-=-=- ###
```

```python
from qiskit.visualization import plot_state_city, plot_bloch_multivector
from qiskit.visualization import plot_state_paulivec, plot_state_hinton
from qiskit.visualization import plot_state_qsphere
```

```python
backend = BasicAer.get_backend('statevector_simulator')
result = execute(circuit, backend).result()
psi = result.get_statevector(circuit)
```

```python
plot_state_city(psi, title="My State City", figsize = (12,5), color=['orange', 'black'])
```

```python
plot_state_hinton(psi, title="My State Hinton", figsize = (5,5))
```

```python
plot_state_qsphere(psi, figsize = (5,5))
```

```python
plot_state_paulivec(psi, title="My State Paulivec", figsize = (8,3), color=['green', 'blue', 'purple'])
```

```python
plot_bloch_multivector(psi, title="My Bloch Multivector", figsize = (8,3))
```

```python
### -=-=-=-=- Interactive State Plots for Jupyter Notebooks -=-=-=-=- ###
```
